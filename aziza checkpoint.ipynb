```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://aziza.tn/fr/home"
page = requests.get(url)
if page.status_code != 200:
    print("srat erreur")
    exit()
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup 
```

```python
list_products = soup.find(id="list_products-aN").find_all("li")[1:]
len(list_products)
```

```python
product_details = [item.find(class_="product-item-details")for item in list_products]
product_details 
```

```python
product_dict = {"product":[], "price":[]}
for item in product_details:
    product_name = item.find("h3").find("a").get_text().strip()
    product_details = item.find("h3").find(class_="more_title").get_text().strip()
    product_brand = item.find("h3").find(class_="brand").get_text().strip()
    product_dict["product"].append(product_name + " " + product_details + " " + product_brand)
    product_price = item.find(class_="price-box price-final_price").find("span").get_text().strip()
    product_dict["price"].append(float(product_price.replace("TND","").strip().replace(",",".")))
```

```python
df = pd.DataFrame(product_dict)
df["Category"] = "Non Alimentaire" 
df
```

```python
list_products2 = soup.find(id="list_products-a").find_all("li")
product_details2 = [item.find(class_="product-item-details")for item in list_products2]
product_dict2 = {"product":[], "price":[]}
for item in product_details2:
    product_name = item.find("h3").find("a").get_text().strip()
    product_details = item.find("h3").find(class_="more_title").get_text().strip()
    product_brand = item.find("h3").find(class_="brand").get_text().strip()
    product_dict2["product"].append(product_name + " " + product_details + " " + product_brand)
    product_price = item.find(class_="price-box price-final_price").find("span").get_text().strip()
    product_dict2["price"].append(float(product_price.replace("TND","").strip().replace(",",".")))
```

```python
product_dict2
```

```python
df1= pd.DataFrame(product_dict2)
df1["Category"] = "Alimentaire" 
df1
```

```python
df.to_csv("aziza1.csv", index=False)
df1.to_csv("aziza2.csv", index=False)
final1 = pd.read_csv("aziza1.csv")
final2 = pd.read_csv("aziza2.csv")

final3 = pd.concat([final1, final2], ignore_index=True)
final3
final3.to_csv("aziza checkpoint.csv", index=False)
```
