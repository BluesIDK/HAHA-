```python
# Define a date range 
start_date = '2024-04-01'
end_date = '2024-04-07'

neows_url = f'https://api.nasa.gov/neo/rest/v1/feed?start_date={start_date}&end_date={end_date}&api_key={API_KEY}'

response_neo = requests.get(neows_url)
neo_data = response_neo.json()

```

```python
neo_data
```

```python
# Extract asteroids data from all dates

asteroids_list = []
for date in neo_data['near_earth_objects']:
    for asteroid in neo_data['near_earth_objects'][date]:
        asteroid_id = asteroid['id']
        name = asteroid['name']
        abs_mag = asteroid['absolute_magnitude_h']
        min_diameter = asteroid['estimated_diameter']['kilometers']['estimated_diameter_min']
        velocity = asteroid['close_approach_data'][0]['relative_velocity']['kilometers_per_second']

        asteroids_list.append({
            'Asteroid ID': asteroid_id,
            'Asteroid name': name,
            'Minimal estimated diameter (km)': min_diameter,
            'Absolute magnitude': abs_mag,
            'Relative velocity (km/s)': velocity
        })
```

```python
#Convert to DataFrame
df_asteroids = pd.DataFrame(asteroids_list)

df_asteroids.head()
```

```python
df_asteroids.to_csv('asteroids_data.csv', index=False)
```
